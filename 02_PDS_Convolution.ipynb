```python
# Create an irregular random 2D signal
x = np.zeros((9, 9))
x[2:7, 2:7] = [[1.0, 0.9, 0.7, 0.9, 1.0],
               [0.8, 0.6, 1.0, 0.6, 0.8],
               [0.7, 1.0, 1.0, 0.8, 1.0],
               [1.0, 0.3, 0.8, 1.0, 0.9],
               [0.9, 0.9, 0.7, 0.9, 1.0]]
print(x)

# Plot it
plt.imshow(x, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# == Q1 ==
# Crea una señal impulso (delta) en 2D. Usa un kernel de tamaño [3x3].
h = np.zeros((3,3))
h[1,1] = 1
print(h)
```

```python
# == Q2 ==
# Comprueba que la convolución con la señal impulso (delta) equivale al
# producto unitario, es decir, y = x = x*k, donde k=delta
# Esta comprobación puedes hacerla de manera práctica usando el siguiente código
```

```python
# Import the convolution function and convolve x and h
from scipy.signal import convolve2d
y = convolve2d(x, h, mode='same', boundary='fill', fillvalue=0)
print(y)

# Plot it
plt.imshow(y, cmap='gray')
plt.axis('off')
plt.show()

# Comprobación AQUÍ (calcula rate)
dif = x - y
rate = 0
for i in range(0,9):
    for j in range(0,9):
        if x[i,j] == y[i,j]:
               rate = rate + 1
rate =(rate / 81)*100
print(f"Rate of pixels with same value in x and y: {rate}%")
```

```python
# En el ejercicio anterior usamos los siguientes parámetros:
# mode='same', boundary='fill', fillvalue=0. Revisa la documentación de convolve2d
# en la página de scipy y verifíca qué pasa si modificamos los valores de los parámetros.
# Concretamente:
# == Q3 == ¿Qué impacto tiene modificar mode? -- Muestra un ejemplo.
print('mode = same')
y = convolve2d(x, h, mode='full', boundary='fill', fillvalue=0)
print(y)
# == Q4 == ¿Qué impacto tiene modificar boundary? -- Muestra un ejemplo.
print('boundary = symm')
y = convolve2d(x, h, mode='same', boundary='symm', fillvalue=0)
print(y)
# == Q5 == ¿Qué impacto tiene modificar fillvalue? -- Muestra un ejemplo.
print('fillvalue = 1')
y = convolve2d(x, h, mode='full', boundary='fill', fillvalue=1)
print(y)
```

```python
# Function to add salt and pepper noise to an image
def salt_and_pepper(IMG):
    img = IMG.copy()
    n_rows, n_cols = img.shape
    for row in range(n_rows):
        for col in range(n_cols):
            thresh = np.random.rand()
            if thresh > 0.95:
                img[row, col] = 1
            elif thresh < 0.05:
                img[row, col] = 0
    return img
```

```python
# Create a noisy copy of the image
noisyIMG = salt_and_pepper(IMG)
plt.imshow(noisyIMG, cmap='gray')
plt.axis('off')
plt.show()
print(f"Average intensity: {noisyIMG.mean():4.2f}+-{noisyIMG.std():4.2f}")
print(f"Min and max pixel values = [{noisyIMG.min()}, {noisyIMG.max()}]")
print(f"There is {100 * np.abs((IMG - noisyIMG)).sum() / (IMG.shape[0] * IMG.shape[1]):4.2f}% of pixels average variation.")
```

```python
# Create a Gaussian kernel and use it to remove the noise
kernel = cv2.getGaussianKernel(3, 0.5)
Gk = np.outer(kernel, kernel.transpose())
Go = convolve2d(noisyIMG, Gk, mode='same', boundary='fill', fillvalue=0)

# Plot convolved image
plt.imshow(Go, cmap='gray')
plt.axis('off')
plt.show()
print(f"There is {100 * np.abs((Go - noisyIMG)).sum() / (Go.shape[0] * Go.shape[1]):4.2f}% of pixels average variation.")
```

```python
# Border/Edge detection
from scipy import misc
IMG2 = misc.ascent()

# Plot convolved image
plt.imshow(IMG2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Sobel line detector
Sobk = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
print(Sobk)
Soby = convolve2d(IMG2, Sobk.transpose(), mode='same', boundary='fill', fillvalue=0)
print(f"Min and max pixel values = [{Soby.min()}, {Soby.max()}]")

# Plot convolved image
plt.imshow(np.abs(Soby), cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Bueno, resulta que el Laplaciano es muy sensible al ruido, por lo que normalmente
# la imágen es suavizada primero mediante un filtro Gaussiano, y el Laplaciano se aplica
# sobre la imagen ya suavizada
```

```python
# Laplacian border detector: Gaussian + Laplacian
Gauk = cv2.getGaussianKernel(5, 1)
Gauk = np.outer(Gauk, Gauk.transpose())
Gauy = convolve2d(IMG2, Gauk, mode='same', boundary='fill', fillvalue=0)

Lapk = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
Lapy = convolve2d(Gauy, Lapk, mode='same', boundary='fill', fillvalue=0)

# Plot convolved image
plt.imshow(np.abs(Lapy), cmap='gray')
plt.axis('off')
plt.show()
```
